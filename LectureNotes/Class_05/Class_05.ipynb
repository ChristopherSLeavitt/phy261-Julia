# Class 05: Electric field of a finite line charge on axis


Let's compute the electric field on the axis of a finite line charge (charge = Q, length = L).
The geometry looks like this:


```julia
ϵ₀ = 8.8541878e-12
```

```julia
function E_stick(x, N, L=0.10, Q=1.0e-9)
    Δx = L/N
    E = 0.0 # starting value for sum
    for i in range(1,N)
        E = E + (1/(4π*ϵ₀))*(Q*Δx/L)/(x - 0.5*(2*i-1)*Δx)^2
    end
    return E
end
    
```

```julia
E_stick(1.0, 10, 0.10 ,0.0)
```

```julia
E_stick(1.0, 10, 0.001)
```

```julia
E(x,Q,L) = (Q/L)/(4π*ϵ₀) * (1.0/(x-L) -1/x)
```

```julia
E(1.0, 1.0e-9, 0.10)
```

```julia
E_stick(1.0,100,0.10,1.0e-9)
```

```julia
x = LinRange(0.101, 1.0, 1000)
```
